## **For: Bucles controlados por variable contadora**

ControlContador.png




En este tipo de bucles la ejecución repetida de una instrucción o secuencia de instrucciones depende de una variable que se toma como índice o contador de forma que cuando esta alcanza un determinado límite la ejecución se para.

Antes de la ejecución de la primera instrucción contenida en el bucle se evalúa la condición y si esta se evalúa a `True`, es decir, se cumple, entonces el programa ejecuta el código contenido en este bloque.

La sintaxis de esta estructura responde al siguiente esquema:

```
  for variable_indice in range(tamaño):
    instruccion_1
    instruccion_2
    ...
```

La función range crea un rango de valores e inicializa la variable_indice a 0 si no se especifica un límite inferior, actualizando su valor en cada iteración y estableciendo el valor de tamaño-1 como límite superior. 

Veámoslo con un ejemplo:

````python
'''
  El siguiente código crea un bucle en el que la variable i solo existe
  en este bloque y además se inicializa con el valor 0.
  En cada iteración sobre esta estructura se incrementa el valor de i automáticamente
  en 1 (como si hiciéramos i+=1) hasta que alcanza el límite 5-1 --> 4, momento
  en que se da la condición de salida del bucle.
  Como puede observarse, en cada iteración se imprime el valor que tiene la variable i
  en ese momento.
'''
for i in range(5):
  print(i)

# Recuerda que el nombre 'i' de variable es habitual como índice en los bucles,
# arrays y listas, pero podemos darle el nombre que queramos.
````

La función range nos permite crear diferentes tipos de rango sobre los que iterar.
Su sintaxis es `range(limite_inferior, limite_superior, incremento)`, pero también nos permite llamarla o ejecutarla sin pasar esos tres valores (como hicimos en el ejemplo anterior).

Cuando solo se especifica un valor la función range entiende que hemos especificado solo el límite superior.

```
  range(3)   #entiende que va de 0 a 2, siendo 3 el límite superior 
```


Sin embargo, si incluimos dos valores asigna el primero al límite inferior y el segundo al superior:

```
  range(1, 6)   # entiende que va de 1 a 5 
```

Finalmente, sin incluimos tres valores en la llamada, interpreta que el primero es el límite inferior, el segundo el superior y el tercero el incremento que se realiza sobre el índice en cada paso:

```
  range(1, 16, 2)   # entiende que va de 1 a 15 de dos en dos 
```

````python
# Va de 0 a 2 de uno en uno
for indice in range(3):
  print(indice)
````

````python
# Va de 2 a 6 de uno en uno
for indice in range(2, 7):
  print(indice)
````

````python
# Va de 5 a 19 de tres en tres
for indice in range(5, 20, 3):
  print(indice)
````

## **While: Bucles controlados por condición**

Captura de pantalla de 2022-10-01 20-14-55.png

Los bucles de tipo while son otro tipo de estructuras iterativas que no tienen obligatoriamente estar asociados a una variable índice o contador (aunque pueden estarlos).

Al igual que los anteriores su ejecución depende del cumplimiento de una condición o expresión lógica que se escribe a continuación de la palabra reservada **while** y antes del símbolo **:**

Su estructura es la siguiente:

```
  while expresion_logica :
    sentencia_1
    sentencia_2
    ...

```

El grupo de sentencias puede no ejecutarse, ejecutarse al menos en una ocasión o múltiples ocasiones, dependiendo del cumplimiento o no de la `expresion_logica`.

Veámoslo con un ejemplo:


````python
edad = 15

while edad < 18:
  print("Tiene acceso al bono de 400€ para gastos culturales, ¡enhorabuena!")
  edad = int(input("Introduzca su edad: "))

print("Es demasiad@ viej@ para acceder a este bono; ya deberías pagarte tú el cine.")
````

````python
# Sin embargo, si inicializamos la variable a un valor superior a 18 ocurre lo siguiente:

edad = 18

while edad < 18:
  print("Tiene acceso al bono de 400€ para gastos culturales, ¡enhorabuena!")
  edad = int(input("Introduzca su edad: "))

print("Es demasiad@ viej@ para acceder a este bono; ya deberías pagarte tú el cine.")

# La condición del bucle no se cumple y no se ejecutan nunca las instrucciones que contiene
````

**Un bucle for se puede convertir a un bucle while haciendo uso de una variable contadora.**

No obstante, en el bucle while debemos obligatoriamente actualizar la variable contadora, porque en caso contrario, se convertiría en un bucle sin condición de parada y no terminaría nunca.

Cuando esto ocurre normalmente nuestro IDE se queda como bloqueado, pero realmente esta ejecutando la instrucción contenida en el bucle sin parar. Podemos comprobar esto ejecutando en modo de depuración y poniendo un punto de interrupción o breakpoint en una de las instrucciones interiores.

````python
for variable in range(5):
  print(variable)
````

````python
contadora = 0

while contadora < 5:
  print(contadora)
  contadora+=1
````

````python
# Por el contrario, este bucle no terminaría nunca e imprimiría 0 constantemente

contadora = 0

while contadora < 5:
  print(contadora)
````

## **Validación de datos**

Los bucles nos pueden ayudar a validar los datos que introduce el usuario, volviendo a pedir que introduza la información cuando esta no es válida, por ejemplo:

````python
from datetime import date


edad = int(input("Introduzca su edad: "))

while edad < 0 or edad > 120:
  edad = int(input("Si puede ser introduzca mejor una edad válida [0-120]: "))

print(f"Naciste en el año {date.today().year-edad}")
````

A continuación podemos ver un ejemplo de menú que nos permite ingresar y sacar dinero de una cuenta realizando algunas validaciones sobre las cantidades que ingresa o extrae el usuario y un resultado de ejecución.

````python
MENU_PRINCIPAL = '''
          Menú principal:
          ---------------
            1. Ingresar
            2. Sacar
            3. Consultar balance
            4. Imprimir menú
            5. Salir
      '''
  
print(MENU_PRINCIPAL)
opcion = int(input("Introduzca una opción: "))
total = 0

while opcion != 5:

  if opcion==1:
    cantidad = int(input("Introduzca la cantidad a ingresar: "))
    while cantidad < 0: 
      cantidad = int(input("Introduzca una cantidad válida: "))
    total +=cantidad

  elif opcion==2:
    cantidad = int(input("Introduzca la cantidad a sacar: "))
    while cantidad < 0 and cantidad > total: 
      cantidad = int(input("Introduzca una cantidad válida: "))
    total -=cantidad
  
  elif opcion==3:
    print(f"El saldo actual es de {total}€")

  elif opcion==4:
    print(MENU_PRINCIPAL)
  
  print("\n\n***********************")
  opcion = int(input("Introduzca una opción: "))


````
